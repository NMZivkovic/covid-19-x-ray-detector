```python
import pandas as pd
import shutil
```

```python
metadata = pd.read_csv('./covid-chestxray-dataset/metadata.csv')
metadata.head()
```

```python
images = metadata[(metadata['finding'] == 'COVID-19') & (metadata['view'] == 'PA')]
images.reset_index()
images.shape
```

```python
for _, row in images.iterrows():
    shutil.copy(f"./covid-chestxray-dataset/images/{row.filename}", f"./data/{row.filename}")
```
